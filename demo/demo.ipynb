# 2. MES(manufacturing execution system) 执行系统/工单派发系统

Integrate the MES system with the APS system to achieve real-time production scheduling and monitoring. 将 MES 系统与 APS 系统集成，实现实时生产调度和监控。

APS =Forecasting and Planning=> MES =Scheduling and Monitoring=> Production Quality Control/Check=> Finished Product For APS/MES System to evaluate
- (Plan Creation in APS) 在 APS 中创建计划
- (Plan Transmission to MES) 计划传输到 MES
- (Execution in MES) 在 MES 中执行
- (Feedback to APS) 反馈至 APS

## Key Features
- Real-time monitoring and tracking
- Real-time scheduling and management
- Real-time data collection and analysis

## Algorithm
Similar to the planning part of APS system. The difference is that the MES system is more focused on the real-time scheduling and monitoring of the production process.

Consider adding a dynamic dispatching algorithm to the MES system to improve the efficiency of the production process.(e.g. Random Forest, Neural Network, uber dispatching algorithm, etc.)
考虑在 MES 系统中加入动态调度算法，以提高生产过程的效率（如随机森林、神经网络、打车调度算法等）。

# Simulated annealing (SA)

Let s = s0
For k = 0 through kmax (exclusive):
T ← temperature( 1 - (k+1)/kmax )
Pick a random neighbour, snew ← neighbour(s)
If P(E(s), E(snew), T) ≥ random(0, 1):
s ← snew
Output: the final state s
